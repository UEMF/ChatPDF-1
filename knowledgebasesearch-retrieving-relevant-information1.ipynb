```python
from langchain.embeddings import HuggingFaceEmbeddings
from huggingface_hub import hf_hub_download
```

```python
hf_embeddings = HuggingFaceEmbeddings(model_name='')
```

```python
import pinecone
```

```python
index_name = ''

pinecone.init(
        api_key="",  # find api key in console at app.pinecone.io
        environment=""  # find next to api key in console
)

# # we create a new index
 pinecone.create_index(
         name=index_name,
         metric='cosine',
         dimension=768 
  )
```

```python
index = pinecone.Index(index_name)
```

```python

batch_limit = 100
metadatas = []

with open(path) as json_file:
    data = json.load(json_file)
    for item in data:
        metadata = {
        'source': item['filename'],
        'page_number': item['page_number'],
        'content':item['content']
            }
        metadatas.append(metadata)

json_file.close()


        
```

```python
index.describe_index_stats()
```

```python
from langchain.vectorstores import Pinecone
vectorstore = Pinecone(index,hf_embeddings.embed_query, [t['content'] for t in metadatas])
```

```python
docsearch = Pinecone.from_texts([t['content'] for t in metadatas], hf_embeddings, index_name=index_name)
```

```python
query = "Qui est Ahmed al-Mansour?"
docs = docsearch.similarity_search(query,k=1)
```

```python
for t in metadatas:
    if docs[0].page_content==t['content']:
        print(docs[0].page_content,t['source'],t['page_number'])
```
